```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns",None)
```

```python
from data_processing import clean_and_process_data

path = "API_ST.INT.ARVL_DS2_en_csv_v2_5994899.csv"

travel = clean_and_process_data(path)
travel = travel.drop(columns=['Country_Code'])
travel.head()
```

```python
countries = travel.Country_Name.values
```

```python
travel.iloc[:,2:].head()
```

```python
time_data_numeric = travel.drop(columns=["Country_Name"]).apply(pd.to_numeric, errors='coerce')
result = time_data_numeric.apply(lambda x: np.sqrt(x), axis=0).round(2)
result.head()
```

```python
data = pd.concat([travel[["Country_Name"]], result],axis=1)
data.head()
```

```python
data["Tourist_Arrivals"] = data.drop(columns=["Country_Name"]).sum(axis=1)
data.head()
```
